```python
def lagr(x, val, t):
    l = 0
    for i in range(len(val)):
        num = 1
        den = 1
        for j in range(len(x)):
            if i == j:
                num *= 1
                den *= 1
            else:
                num *= (t - x[j])
                den *= (x[i] - x[j])
        l += val[i] * num / den
    return l
```

```python
#Вспомогательные функции

from math import factorial

def coef(m, n):
    return factorial(m)/(factorial(m-n)*factorial(n))

def kr(y, i, k):#Конечная разность
    if k == 0:
        return y[i]
    else:
        S = 0
        for j in range(k+1):
            S += (-1)**j*coef(k, j)*y[k+i-j]
        return S

def kr2(y, i, k):#Разница разностей
    if k == 0:
        return y[i]
    else:
        return kr2(y, i+1, k-1) - kr2(y, i, k-1)
```

```python
def newton_one(x, X, Y):
    P = 0
    q = 1
    
    for i in range(len(Y)):
        P += kr2(Y, 0, i) * q / factorial(i)
        q *= x - X[i]
    return P
```

```python
def newton_two(x, X, Y):
    P = Y[-1]
    q = x - X[-1]
    j = 1
    
    for i in range(len(Y)-2, -1, -1):
        P += kr2(Y, i, j)*q/factorial(j)
        q *= x - X[i]
        j += 1
    return P
```

```python
def c(x, k):
    s = 0
    for i in range(len(x)):
        s += x[i]**k
    return s

def b(x, y, k):
    s = 0
    for i in range(len(x)):
        s += y[i]*x[i]**k
    return s

def a( X, Y):
    m = 4
    
    C = np.zeros((m, m))
    B = np.zeros((m))
    A = np.zeros((m))
    
    for i in range(m):
        l = 0
        for j in range(i, m+i):
            C[i][l] = c(X, j)
            l += 1
    
    for i in range(m):
        B[i] = b(X, Y, i)
        
    A = np.linalg.solve(C, B)
    return A
    
def approx(x,A):
    m=4
    P = A[0]
    for i in range(1, m):
        P += A[i]*x**i
    return P
```

```python
n = 12
def cubic_interp1d(x0, x, y):
    x = np.asfarray(x)
    y = np.asfarray(y)

    if np.any(np.diff(x) < 0):
        indexes = np.argsort(x)
        x = x[indexes]
        y = y[indexes]

    size = len(x)

    xdiff = np.diff(x)
    ydiff = np.diff(y)


    Li = np.empty(size)
    Li_1 = np.empty(size-1)
    z = np.empty(size)


    Li[0] = (2*xdiff[0])**(0.5)
    Li_1[0] = 0.0
    B0 = 0.0 
    z[0] = B0 / Li[0]

    for i in range(1, size-1, 1):
        Li_1[i] = xdiff[i-1] / Li[i-1]
        Li[i] = (2*(xdiff[i-1]+xdiff[i]) - Li_1[i-1] * Li_1[i-1])**0.5
        Bi = 6*(ydiff[i]/xdiff[i] - ydiff[i-1]/xdiff[i-1])
        z[i] = (Bi - Li_1[i-1]*z[i-1])/Li[i]

    i = size - 1
    Li_1[i-1] = xdiff[-1] / Li[i-1]
    Li[i] = (2*xdiff[-1] - Li_1[i-1] * Li_1[i-1])**0.5
    Bi = 0.0
    z[i] = (Bi - Li_1[i-1]*z[i-1])/Li[i]

    i = size-1
    z[i] = z[i] / Li[i]
    for i in range(size-2, -1, -1):
        z[i] = (z[i] - Li_1[i-1]*z[i+1])/Li[i]

    index = x.searchsorted(x0)
    np.clip(index, 1, size-1, index)

    xi1, xi0 = x[index], x[index-1]
    yi1, yi0 = y[index], y[index-1]
    zi1, zi0 = z[index], z[index-1]
    hi1 = xi1 - xi0

    f0 = zi0/(6*hi1)*(xi1-x0)**3 + \
         zi1/(6*hi1)*(x0-xi0)**3 + \
         (yi1/hi1 - zi1*hi1/6)*(x0-xi0) + \
         (yi0/hi1 - zi0*hi1/6)*(xi1-x0)
    return f0
```

```python
x4 = np.arange(n)
y4 = val4[:n]

x41 = np.arange(0, n - 1 + 0.1, 0.1)
y41 = cubic_interp1d(x41, x4, y4)

plt.figure(figsize=(15, 7))
plt.plot(x4, y4, "rx", x41, y41)
plt.grid()
plt.title("Интерполяция кубическим сплайном 12 значений температур в феврале, шаг 0.1")
plt.show()
```

```python
N = np.array(frame['фев'])   
m = int(len(N)/2-1)     

def A(k, x, y):
    A = 0
    n = len(x)
    for i in range(n):
        A += y[x[i]]*np.cos(k*2*np.pi*i/n)
    return 2*A/n

def B(k, x, y):
    A = 0
    n = len(x)
    for i in range(n):
        A += y[x[i]]*np.sin(k*2*np.pi*i/n)
    return 2*A/n

    
def approx(x, X, Y, arrA, arrB, m=None):
    N = len(X)
    if m == None:
        m = int(N/2 - 1)
        
    F = 1/N*sum(Y)
    for k in range(1,m+1):
        F += arrA[k-1]*np.cos(2*np.pi*k*x/N) + arrB[k-1]*np.sin(2*np.pi*k*x/N)
    return F
```

```python
x = np.arange(len(N))
y = N

arrA = [A(k, x, y) for k in range(1, m+1)]
arrB = [B(k, x, y) for k in range(1, m+1)]

x1 = np.arange(0, len(N), 1)
y1 = [approx(i, x, y, arrA, arrB, m) for i in x1]

plt.figure(figsize=(15, 7))
plt.plot(x, y, "o", ms=0.5, label="Values 'Фев'")
plt.plot(x1, y1, "--", lw=2, label="Approximation")
plt.grid()
plt.xlim(0, len(N))
plt.title("Аппроксимация тригонометрическими функциями")
plt.legend()
plt.show()
```
